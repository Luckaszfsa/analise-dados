```python
import pandas as pd

# carregar os dados do arquivo CSV
dados = pd.read_csv(r'C:\Users\User\Projetos an√°lise de dados\analise-dados\Projetos\Futebol\BRA.csv')

# criar fun√ß√£o para definir resultados de cada jogo
def definir_resultado(linha):
    if linha['Res'] == 'H':
        return 'Vitoria'
    elif linha['Res'] == 'A':
        return 'Derrota'
    else:
        return 'Empate'
# criar fun√ß√£o para calcular a pontua√ß√£o
def calcular_pontos(linha):
    return linha['Vitoria'] * 3 + linha['Empate']



# agrupar os dados por time e resultado e calcular as contagens
resultados_por_time = dados.groupby(['Home', 'Resultado']).size().reset_index(name='Contagem')

# pivotar a tabela para ter uma coluna para cada tipo de resultado
resultados_por_time = resultados_por_time.pivot(index='Home', columns='Resultado', values='Contagem')

# preencher valores NaN com zero
resultados_por_time = resultados_por_time.fillna(0)

# calcular o n√∫mero total de jogos de cada time
resultados_por_time['Total'] = resultados_por_time['Vitoria'] + resultados_por_time['Derrota'] + resultados_por_time['Empate']
# aplicar fun√ß√£o para criar coluna "Pontos"
resultados_por_time['Pontos'] = resultados_por_time.apply(calcular_pontos, axis=1)


# salvar o resultado em um arquivo CSV
resultados_por_time.to_csv('resultados_por_time.csv')
```

```python
import pandas as pd

#carregar os dados do arquivo CSV
dados = pd.read_csv(r'C:\Users\User\Projetos an√°lise de dados\analise-dados\Projetos\Futebol\BRA.csv')

#criar fun√ß√£o para definir resultados de cada jogo
def definir_resultado(linha):
    if linha['Res'] == 'H':
        return 'Vitoria'
    elif linha['Res'] == 'A':
        return 'Derrota'
    else:
        return 'Empate'

#aplicar fun√ß√£o para criar coluna "Resultado"
dados['Resultado'] = dados.apply(definir_resultado, axis=1)

#agrupar os dados por time e resultado e calcular as contagens
resultados_por_time = dados.groupby(['Home', 'Resultado']).size().reset_index(name='Contagem')

#pivotar a tabela para ter uma coluna para cada tipo de resultado
resultados_por_time = resultados_por_time.pivot(index='Home', columns='Resultado', values='Contagem')

#preencher valores NaN com zero
resultados_por_time = resultados_por_time.fillna(0)

#calcular o n√∫mero total de jogos de cada time
resultados_por_time['Total'] = resultados_por_time['Vitoria'] + resultados_por_time['Derrota'] + resultados_por_time['Empate']

#calcular a pontua√ß√£o de cada time
resultados_por_time['Pontos'] = resultados_por_time['Vitoria'] * 3 + resultados_por_time['Empate']

#ordenar o DataFrame pela pontua√ß√£o de forma decrescente
resultados_por_time = resultados_por_time.sort_values(by='Pontos', ascending=False)

#salvar o resultado em um arquivo CSV
resultados_por_time.to_csv('resultados_por_time.csv')
```

```python
import pandas as pd
import sqlite3

# carregar os dados do arquivo CSV
dados = pd.read_csv(r'C:\Users\User\Projetos an√°lise de dados\analise-dados\Projetos\Futebol\BRA.csv')

# criar fun√ß√£o para definir resultados de cada jogo
def definir_resultado(linha):
    if linha['Res'] == 'H':
        return 'Vitoria'
    elif linha['Res'] == 'A':
        return 'Derrota'
    else:
        return 'Empate'

# aplicar fun√ß√£o para criar coluna "Resultado"
dados['Resultado'] = dados.apply(definir_resultado, axis=1)

# agrupar os dados por time e calcular as estat√≠sticas
gols_por_time = dados.groupby(['Home']).agg({'HG': 'sum', 'AG': 'sum'}).reset_index()
gols_sofridos_por_time = dados.groupby(['Home']).agg({'AG': 'sum', 'HG': 'sum'}).reset_index()
resultados_por_time = dados.groupby(['Home', 'Resultado']).size().reset_index(name='Contagem')
resultados_por_time = resultados_por_time.pivot(index='Home', columns='Resultado', values='Contagem').fillna(0)
resultados_por_time['Total'] = resultados_por_time['Vitoria'] + resultados_por_time['Derrota'] + resultados_por_time['Empate']
resultados_por_time['Pontos'] = resultados_por_time['Vitoria']*3 + resultados_por_time['Empate']

# juntar as informa√ß√µes em um √∫nico dataframe
dados_agrupados = gols_por_time.merge(gols_sofridos_por_time, on='Home', suffixes=['_feitos', '_sofridos'])
dados_agrupados = dados_agrupados.merge(resultados_por_time, on='Home')

# criar conex√£o com o banco de dados
conn = sqlite3.connect('dados_futebol.db')

# criar tabela para armazenar os dados
conn.execute('''CREATE TABLE IF NOT EXISTS dados_futebol
                (time TEXT, gols_casa INTEGER, gols_fora INTEGER, total_gols INTEGER, 
                 gols_sofridos_casa INTEGER, gols_sofridos_fora INTEGER, total_gols_sofridos INTEGER, 
                 vitorias INTEGER, empates INTEGER, derrotas INTEGER, total_jogos INTEGER, pontos INTEGER)''')

# inserir os dados na tabela
for row in dados_agrupados.itertuples(index=False):
    conn.execute(f"INSERT INTO dados_futebol VALUES ('{row.Home}', {row.HG_feitos}, {row.AG_feitos}, {row.HG_feitos + row.AG_feitos}, {row.HG_sofridos}, {row.AG_sofridos}, {row.HG_sofridos + row.AG_sofridos}, {row.Vitoria}, {row.Empate}, {row.Derrota}, {row.Total}, {row.Pontos})")

# commitar as mudan√ßas e fechar a conex√£o
conn.commit()
conn.close()
```
