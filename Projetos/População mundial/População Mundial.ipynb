```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
df = pd.read_csv(r'C:\Users\User\Projetos análise de dados\analise-dados\Projetos\População mundial\data_population_world.csv')
```

```python
df.head()
```

```python
df.drop_duplicates()
```

```python
df = df.drop(["Indicator Name","Indicator Code"],axis=1)
```

```python
df.head()
```

```python
df.head()
```

```python
df = df.dropna(subset=['Country Name'])
```

```python
df.head()
```

```python
# Criar um gráfico de linha da população total ao longo do tempo para um país específico
pais = 'Brazil'
pop_por_ano = df.loc[df['Country Name'] == pais, '1960':'2021'].T
pop_por_ano.columns = [pais]
pop_por_ano.plot(kind='line', title='População total do Brasil')
```

```python
# Criar um gráfico de linha da população total ao longo do tempo para vários países específicos
paises = ['Brasil', 'China', 'Estados Unidos']
pop_por_ano = df[df['Country Name'].isin(paises)].set_index('Country Name').loc[paises, '1960':'2021']
pop_por_ano = pop_por_ano.T
pop_por_ano.plot(kind='area', title='População total do Brasil, China e EUA')
```

```python
df.plot(kind='line', legend=False)
plt.title('População total ao longo do tempo')
plt.xlabel('Ano')
plt.ylabel('População')
plt.show()
```

```python
# Selecionar apenas as colunas de anos e aplicar o método sum para cada ano
pop_por_ano = df.iloc[:, 4:].sum()
```

```python
# Plotar um gráfico de linha
pop_por_ano.plot(kind='line', title='População total por ano em bilhões')
```

```python
# Seleciona apenas as colunas que contém a população por ano
pop_por_ano = df.loc[:, '1960':'2021']

# Calcula o percentual de crescimento por ano
pct_crescimento = pop_por_ano.pct_change(axis='columns') * 100

# Exibe o resultado
display(pct_crescimento)
```

```python
pct_crescimento.plot(kind='line', title='Percentual de crescimento anual da população mundial')
```

```python

# calcular a população total por ano
pop_por_ano = df.iloc[:, 4:].sum()

# criar o boxplot
fig, ax = plt.subplots()
ax.boxplot(pop_por_ano, vert=False)

# definir o título e os rótulos dos eixos
ax.set_title('Distribuição da população ao longo dos anos')
ax.set_xlabel('População')
ax.set_yticklabels(['Todos os anos'])

# exibir o gráfico
plt.show()
```
